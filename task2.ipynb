```python
import numpy as np
import pandas as pd
```

```python
data_train = pd.read_csv("train.csv")
```

```python
data_train
```

```python
data_train.shape
```

```python
data_train["dx"] = 0
data_train["dy"] = 0
data_train["dz"] = 0
data_train["dTx"] = 0
data_train["dTy"] = 0
dx = np.array([])
for i in range(500):
    data_loc = data_train[data_train["data_ind"] == i]
    index = data_loc["ele_P"][pd.notnull(data_loc["ele_P"])].index[0]
    data_train.loc[data_train["data_ind"] == i, "dx"] = data_loc["X"] - data_loc["X"][index]
    data_train.loc[data_train["data_ind"] == i, "dy"] = data_loc["Y"] - data_loc["Y"][index]
    data_train.loc[data_train["data_ind"] == i, "dz"] = data_loc["Z"] - data_loc["Z"][index]
    data_train.loc[data_train["data_ind"] == i, "dTx"] = data_loc["TX"] - data_loc["TX"][index]
    data_train.loc[data_train["data_ind"] == i, "dTy"] = data_loc["TY"] - data_loc["TY"][index]
```

```python
data_train
```

```python
data_train_new = data_train.drop('index', axis=1)
train_target = np.array(data_train_new["signal"])
data_train_new = data_train_new.drop('signal', axis=1)
data_train_new = np.array(data_train_new)
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(data_train_new, train_target)
```

```python
data_test = pd.read_csv("new_test.csv")
data_test.shape
```

```python
data_test[pd.notnull(data_test["ele_P"])]
```

```python
data_test = data_test.drop('index', axis=1)
data_test["dx"] = 0
data_test["dy"] = 0
data_test["dz"] = 0
data_test["dTx"] = 0
data_test["dTy"] = 0
for i in range(500):
    data_loc = data_test[data_test["data_ind"] == i]
    index = data_loc["ele_P"][pd.notnull(data_loc["ele_P"])].index[0]
    data_test.loc[data_test["data_ind"] == i, "dx"] = data_loc["X"] - data_loc["X"][index]
    data_test.loc[data_test["data_ind"] == i, "dy"] = data_loc["Y"] - data_loc["Y"][index]
    data_test.loc[data_test["data_ind"] == i, "dz"] = data_loc["Z"] - data_loc["Z"][index]
    data_test.loc[data_test["data_ind"] == i, "dTx"] = data_loc["TX"] - data_loc["TX"][index]
    data_test.loc[data_test["data_ind"] == i, "dTy"] = data_loc["TY"] - data_loc["TY"][index]
```

```python
data_test = np.array(data_test)
predicted = model.predict_proba(data_test)
```

```python
sort_train = data_train.sort_values(["Z", 'data_ind'], ascending=[1, 0])
df_split = np.array_split(sort_train, 4)
```

```python
data_train_split_1 = df_split[0].drop('id', axis=1)
train_target_1 = np.array(data_train_split_1["signal"])
data_train_split_1 = data_train_split_1.drop('signal', axis=1)
data_train_split_1 = np.array(data_train_split_1)
data_test = data_test.drop('id', axis=1)
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(data_train_split_1, train_target_1)
```

```python
predicted_1 = model.predict_proba(np.array(df_split_test[0]))
df_split_test[0]["prob"] = predicted_1[:, 1]
```

```python
data_train_split_2 = df_split[1].drop('id', axis=1)
train_target_2 = np.array(data_train_split_2["signal"])
data_train_split_2 = data_train_split_2.drop('signal', axis=1)
data_train_split_2 = np.array(data_train_split_2)


model_2 = XGBClassifier()
model_2.fit(data_train_split_2, train_target_2)

predicted_2 = model_2.predict_proba(np.array(df_split_test[1]))
df_split_test[1]["prob"] = predicted_2[:, 1]
```

```python
indices = pd.Int64Index(np.arange(0, test_final.shape[0]))
indices = pd.DataFrame(indices)
csv_y_pred = pd.DataFrame(data=test_final)
csv_y_pred = pd.concat([indices, csv_y_pred], axis=1)
csv_y_pred.columns = ['id', 'signal']
csv_y_pred.to_csv('answers_test_stack.csv', index=False)
```
